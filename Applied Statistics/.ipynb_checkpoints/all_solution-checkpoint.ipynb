```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Import required libraries and import csv file into a dataframe.
dataset_path = 'honeyproduction.csv'
df = pd.read_csv(dataset_path)

df.head()
```

```python
# 2) Show the column names of all individual datasets.
print("Column Names:")
print(df.columns.tolist())

```

```python
# 3) Change the datatype of all columns except price per lb.
columns_to_exclude = ['priceperlb', 'state']
for column in df.columns:
    if column not in columns_to_exclude:
        df[column] = pd.to_numeric(df[column], errors='coerce')

df.head()
```

```python
# 4) Share 5-point summary and the details about the dataset.
summary = df.describe()
summary
```

```python
# 5) Find the average production per state and shape of dataset.
average_production_per_state = df.groupby('state')['totalprod'].mean()
dataset_shape = df.shape
print("Average Production per State:")
print(average_production_per_state)
print("\nShape of the Dataset:")
print(dataset_shape)

```

```python
# 6) Check the distribution of records for every year.
yearly_distribution = df['year'].value_counts().sort_index()
print("Distribution of Records for Every Year:")
print(yearly_distribution)

```

```python
# 7) Find top 10 years with highest totalprod.
top_10_years_totalprod = df.groupby('year')['totalprod'].sum().nlargest(10)
print("Top 10 Years with Highest Total Production:")
print(top_10_years_totalprod)
```

```python

# 8) Find years with highest and lowest totalprod.
year_highest_total = df.groupby('year')['totalprod'].sum().idxmax()
year_lowest_total = df.groupby('year')['totalprod'].sum().idxmin()
print("Year with Highest Total Production:", year_highest_total)
print("Year with Lowest Total Production:", year_lowest_total)

```

```python
# 9) Find top 10 states with highest totalprod.
top_10_states_totalprod = df.groupby('state')['totalprod'].sum().nlargest(10)
print("Top 10 States with Highest Total Production:")
print(top_10_states_totalprod)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset_path = 'honeyproduction.csv'
df = pd.read_csv(dataset_path)

# Set style for plots
sns.set(style="whitegrid")

# 1) Visualize Q 1.9 and Q 1.10
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_states_totalprod.values, y=top_10_states_totalprod.index)
plt.title('Top 10 States with Highest Total Production')
plt.xlabel('Total Production')
plt.ylabel('State')
plt.show()

```

```python
# 2) Find states with minimum and max price per lb
min_price_state = df[df['priceperlb'] == df['priceperlb'].min()]['state'].values[0]
max_price_state = df[df['priceperlb'] == df['priceperlb'].max()]['state'].values[0]
print("State with Minimum Price per lb:", min_price_state)
print("State with Maximum Price per lb:", max_price_state)

```

```python
# 3) Visualize Q 2.2
plt.figure(figsize=(10, 6))
sns.distplot(df['yieldpercol'], bins=20, kde=False)
plt.title('Distribution of Honey Yield per Colony')
plt.xlabel('Honey Yield per Colony')
plt.ylabel('Frequency')
plt.show()

```

```python
# 4) Visualize the totalprod with respect to year
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='totalprod', data=df)
plt.title('Total Production with Respect to Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()

```

```python
# 5) Visualize pairplot and share your insights
sns.pairplot(df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']])
plt.suptitle("Pairplot of Numeric Variables", y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np 

# Load the dataset
dataset_path = 'honeyproduction.csv'
df = pd.read_csv(dataset_path)

# Set style for plots
sns.set(style="whitegrid")

# 1) Visualize distribution of numcol, yieldpercol, priceperlb, stocks in one single frame using subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(df['numcol'], ax=axes[0, 0])
sns.histplot(df['yieldpercol'], ax=axes[0, 1])
sns.histplot(df['priceperlb'], ax=axes[1, 0])
sns.histplot(df['stocks'], ax=axes[1, 1])
axes[0, 0].set_title('Distribution of numcol')
axes[0, 1].set_title('Distribution of yieldpercol')
axes[1, 0].set_title('Distribution of priceperlb')
axes[1, 1].set_title('Distribution of stocks')
plt.tight_layout()
plt.show()

```

```python
# 2) Visualize boxplot and confirm if there are any outliers
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.boxplot(x=df['numcol'], ax=axes[0, 0])
sns.boxplot(x=df['yieldpercol'], ax=axes[0, 1])
sns.boxplot(x=df['priceperlb'], ax=axes[1, 0])
sns.boxplot(x=df['stocks'], ax=axes[1, 1])
axes[0, 0].set_title('Boxplot of numcol')
axes[0, 1].set_title('Boxplot of yieldpercol')
axes[1, 0].set_title('Boxplot of priceperlb')
axes[1, 1].set_title('Boxplot of stocks')
plt.tight_layout()
plt.show()

```

```python
# 3) Find Skewness and Kurtosis of complete dataframe
skewness = df.skew()
kurtosis = df.kurtosis()
print("Skewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)

```

```python
# 4) Make a new copy of dataset and use for further analysis
df_copy = df.copy()

df_copy.head()
```

```python
# 5) Try to make the skewed data as normal as possible & visualize the same
# Applying log transformation to positively skewed columns
skewed_columns = ['numcol', 'yieldpercol', 'priceperlb', 'stocks']
df_copy[skewed_columns] = df_copy[skewed_columns].apply(lambda x: x.apply(lambda y: np.log(y) if y > 0 else 0))

# Visualize the transformed data
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(df_copy['numcol'], ax=axes[0, 0])
sns.histplot(df_copy['yieldpercol'], ax=axes[0, 1])
sns.histplot(df_copy['priceperlb'], ax=axes[1, 0])
sns.histplot(df_copy['stocks'], ax=axes[1, 1])
axes[0, 0].set_title('Distribution of log(numcol)')
axes[0, 1].set_title('Distribution of log(yieldpercol)')
axes[1, 0].set_title('Distribution of log(priceperlb)')
axes[1, 1].set_title('Distribution of log(stocks)')
plt.tight_layout()
plt.show()
```
